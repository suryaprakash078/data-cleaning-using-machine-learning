```python
data = pd.read_csv('cicids.csv')
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dropna()
```

```python
avg = df[' Idle Std'].mean()
```

```python
avg
```

```python
df['Active Mean'].fillna(avg, inplace=True)
```

```python
df.to_excel("updated.xlsx",index=False)
```

```python
files.download("updated.xlsx")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Active Mean'],df[' Active Std'],color='green')
plt.suptitle('graph')
plt.xlabel('ssc marks')
plt.ylabel(' Active Std')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(df['Active Mean'],df[' Active Std'],color='green')
plt.suptitle('graph')
plt.xlabel('Active Mean')
plt.ylabel(' Active Std')
```

```python
df_class = df.groupby('Active Mean').sum()
```

```python
df_class
```

```python
df_class
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df_class[' Flow Duration'], labels=df_class.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Flow Duration by Active Mean')
plt.axis('equal')
```

```python
plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)
plt.scatter(df['Active Mean'], df[' Active Std'], color='green')
plt.title('Scatter Plot: Active Mean vs. Active Std')
plt.xlabel('Active Mean')
plt.ylabel(' Active Std')

plt.subplot(1, 2, 2)
top = df_class.head(11).iloc[1:]
plt.pie(top[' Flow Duration'], labels=top.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Flow Duration by Active Mean (Skipping first record, showing next 10)')
plt.axis('equal')

plt.tight_layout()
plt.savefig('dashboard.png')
```
